```python
import ipl_analysis as ipl
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30,
		'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(courses, values, color ='maroon',
		width = 0.4)

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
fig = plt.figure(figsize = (10, 5))
X_axis = np.arange(len(ipl.venue))
# creating the bar plot
plt.bar(X_axis - 0.2, ipl.bat_first_run_rate, 0.4, label = 'bat frist run rate')
plt.bar(X_axis + 0.2, ipl.bat_second_run_rate, 0.4, label = 'bat second run rate')

plt.xticks(X_axis, ipl.venue)
plt.xlabel("T e a m s")
plt.ylabel("Run rate")
plt.title("Bat first & second run rate")
plt.legend()
plt.show()
```
