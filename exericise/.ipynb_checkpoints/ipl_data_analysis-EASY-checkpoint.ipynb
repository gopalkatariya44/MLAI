```python
import pandas as pd
```

```python
df=pd.read_excel("All_IPL_Match_Scores_raw.xlsx")
df.head()
```

```python
cl=df.groupby('Year')
cl
```

```python
ans=cl.Match_Number.count()
ans
```

```python
tmp=df[df['Winning_Team']=='FirstBatting']
tmp
```

```python
z1=tmp.groupby('Year')
z1.Match_Number.count()
```

```python
ans=pd.concat([ans,z1.Match_Number.count()],axis=1)
ans
```

```python
z2=df[df['Winning_Team']=='Chasing'].groupby('Year')
ans=pd.concat([ans,z2.Match_Number.count()],axis=1)
ans
```

```python
z3=df[df['Winning_Team']=='Match Tied'].groupby('Year')
ans=pd.concat([ans,z3.Match_Number.count()],axis=1)
ans
```

```python
ans=ans.fillna(0,downcast='infer')
ans.columns=['Total Number of Matches','Number of matches Team Batting First Won','Number of matches team Chasing Won','Number of times Match Tied']
ans
```
